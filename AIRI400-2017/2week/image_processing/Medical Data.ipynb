```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
import dicom
```

```python
all_data = []
for i in range(10):
    PathDicom = "./image/data/"+str(i+1)+"/dcm"
    lstFilesDCM = []  # create an empty list
    for dirName, subdirList, fileList in os.walk(PathDicom):
        for filename in fileList:
            if ".dcm" in filename.lower():  # check whether the file's DICOM
                # files url
                files = os.path.join(dirName,filename)
                # read dcm files
                RefDs = dicom.read_file(files)
                # read pixel array from dcm
                lstFilesDCM.append(RefDs.pixel_array)
        all_data.append(lstFilesDCM)
a = np.asarray(all_data)
a.reshape()
```
