# Slicing image

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./image/Frames.jpg', 0)
img3 = cv2.imread('./image/Frames.jpg', 1)
contours_approx = np.zeros(img.shape, np.uint8)
ret, binary = cv2.threshold(img, 220, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(binary, cmap='gray', interpolation = 'bicubic')
plt.show()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
```

```python
plt.imshow(opening, cmap='gray', interpolation = 'bicubic')
plt.show()
```

```python
morph = cv2.morphologyEx(opening, cv2.MORPH_GRADIENT, kernel)
_, contours, hierarchy = cv2.findContours(opening,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
cnt = 1
for box in contours:
    epsilon = 0.08 * cv2.arcLength(box, True)
    approx = cv2.approxPolyDP(box, epsilon, True)
    if len(approx) == 4 :
        area = cv2.contourArea(approx)
        if area > 3000 and area < 10000:
            cv2.drawContours(contours_approx, [approx], -1, 255, 1)
            approx = np.reshape(approx, (4,2))
            
            minY = np.min(approx[:,1])
            minX = np.min(approx[:,0])
            maxY = np.max(approx[:,1])
            maxX = np.max(approx[:,0])
            
            sub = img3[minY:maxY, minX:maxX]
            
#             cv2.imwrite('./images/image{}.png'.format(cnt), sub)
            cnt += 1
```

```python
plt.imshow(contours_approx, cmap='gray', interpolation = 'bicubic')
plt.show()
```
